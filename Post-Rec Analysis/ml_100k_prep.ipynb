__Pre-process original `MovieLens-100K` dataset__

```python
import pandas as pd 
import numpy as np
```

_Dataloading_

```python
# processing user data
user_df = pd.read_csv('../data/ml_100k/u.user', sep='|', header=None, names=['id', 'age', 'gender', 'occupation', 'zip_code'], engine='python',encoding="ISO-8859-1")
user_df.head()
```

```python
user_df.to_csv('../data/ml_100k/user_data.csv', index=False)
```

```python
# processing item data
GENRES = ['unknown', 'Action', 'Adventure', 'Animation',
     'Children', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy',
     'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi',
     'Thriller', 'War', 'Western']
item_df = pd.read_csv('../data/ml_100k/u.item', sep='|', header=None,
                                          names=['id', 'title', 'release_date', 'video_release_date', 'url'] + GENRES,
                                          engine='python',encoding="ISO-8859-1")
item_df.head()
```

```python
item_df.to_csv('../data/ml_100k/item_data.csv', index=False)
```

```python
# processing training data
rating_df = pd.read_csv(
            '../data/ml_100k/u1.base', sep='\t', header=None,
            names=['user_id', 'item_id', 'rating', 'timestamp'],
            dtype={'user_id': np.int32, 'movie_id' : np.int32,
                   'ratings': np.float32, 'timestamp': np.int64}, engine='python',encoding="ISO-8859-1")
rating_df.head()
```

```python
rating_df.to_csv('../data/ml_100k/train_data.csv', index=False)
```

```python
# processing validation data supplied
test_df = pd.read_csv(
            '../data/ml_100k/u1.test', sep='\t', header=None,
            names=['user_id', 'item_id', 'rating', 'timestamp'],
            dtype={'user_id': np.int32, 'movie_id' : np.int32,
                   'ratings': np.float32, 'timestamp': np.int64}, engine='python',encoding="ISO-8859-1")
test_df.head()
```

```python
test_df.to_csv('../data/ml_100k/test_data.csv', index=False)
```
