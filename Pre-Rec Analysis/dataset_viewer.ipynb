Music Preferences in car [Resource for Context Aware Recommendation] \
data source : [link](https://171e731f-93f4-4c62-9b58-c8f419ba01f7.filesusr.com/archives/be17b7_65152091a93140b98d4c9723127521ec.zip?dn=Music_InCarMusic.zip)

```python
rating_data = pd.read_excel('../data/datasets/Data_InCarMusic.xlsx',sheet_name='ContextualRating')
track_data = pd.read_excel('../data/datasets/Data_InCarMusic.xlsx',sheet_name='Music Track')
cat_data = pd.read_excel('../data/datasets/Data_InCarMusic.xlsx',sheet_name='Music Category', header=None)
```

```python
track_data['category'] = track_data[' category_id'].map(dict(zip(cat_data[0], cat_data[1])))
track_data = track_data.loc[:, ['id', ' album', ' artist', ' title',  'category']]
track_data.rename(columns={' album': 'album', ' title': 'title', ' artist': 'artist'})
full_data = pd.merge(left=rating_data, right=track_data, how='inner', left_on='ItemID', right_on='id', )
```

```python
full_data.head()
```

```python
full_data.to_csv('../data/datasets/music_pref.csv', index=False)
```

ReRex, tourism rating dataset \
data source : [link](https://171e731f-93f4-4c62-9b58-c8f419ba01f7.filesusr.com/archives/be17b7_74196b583fc844eab4b9443ae82427b5.zip?dn=rerex.zip)

```python
import glob 

dfs = []
for path in glob.glob('../data/yahoo/*.csv'):
    df = pd.read_csv(path, header=None)
    dfs.append(df)

full_df = pd.concat(dfs, ignore_index=True)
full_df = full_df.iloc[:, :-1]
full_df.columns = ['UserId', 'LoungeId', 'rating', 'visit', 'birthDay', 'gender', 'ageCategory']
full_df.head()
```

```python
full_df.to_csv('../data/datasets/tourist_rating.csv', index=False)
```
